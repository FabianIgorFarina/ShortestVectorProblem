# Examples of SVP as QUBO

```python
import svp_qubo as svp
from sympy import *
init_printing(use_unicode=True)
```

First of all we need to define bad basis vectors $\{\mathbf{b}_1,\dots,\mathbf{b}_{dim-lattice}\}$ of the lattice. Either sympy's Matrix-type or just a list of lists are fine.

```python
lattice = Matrix([[2, 3], [2, 2]])
svp.calculate_gram_matrix(lattice)
```

Since we are looking for the shortest non-zero vector of this lattice, we have to find the minimal (non-zero) value of the vector length (squared):
$$\|\mathbf{x}\|^2 = \sum\limits_{i,j}x_i x_j \mathbf{b}_i\cdot\mathbf{b}_j =: \sum\limits_{i,j} G_{ij} x_i x_j$$

One challenge now becomes that the $x_i$ are whole numbers and with the qubits we only have two energy levels available. Thus we need to map the problem into one, which can be handled by a quantum computer.

There are different ways of doing this, but one intuitive way would be to use the binary number system to encode our problem.

## Sampling

The next function does all the work for us and samples through all possible states of our quantum system and returns the results as a pandas DataFrame and uses binary coding for this.

```python
df = svp.sample_qubits_binary(lattice)
```

We have to sort this now by its energy values.

```python
df.sort_values('energy').head(20)
```

If we could harnest the computational complexity of qutrits instead of qubits, we would need much less of them in order to get the same results. 

```python
df = svp.sample_qutrits_ternary(lattice)
```

```python
df.sort_values('energy').head(20)
```

## Information about our system

### Binary encoding

```python
svp.qubo_coefficients_binary(lattice) # constant [0], linear [1] and quadratic [2] terms of the hamiltonian
```

```python
svp.qubo_hamiltonian_binary(lattice) # the variables and the corresponding hamiltonian
```

### Hamming Encoding

```python
svp.qubo_coefficients_binary(lattice)
```

```python
svp.qubo_hamiltonian_binary(lattice)
```

### Ternary Encoding (partially Hamming)

```python
svp.qubo_coefficients_ternary(lattice)
```

```python
svp.qubo_hamiltonian_ternary(lattice)
```

### Real Ternary with Qutrits

```python
svp.qubo_coefficients_real_ternary(lattice)
```

```python
svp.qubo_hamiltonian_real_ternary(lattice)
```
